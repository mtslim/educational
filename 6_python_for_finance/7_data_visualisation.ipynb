__Static 2D plotting__


```python
import matplotlib as mpl

mpl.__version__
```

```python
import matplotlib.pyplot as plt # import the main plotting (sub)package

plt.style.use('seaborn')

mpl.rcParams['font.family'] = 'serif'

%matplotlib inline
```

__One-dimensional data sets__

```python
#
import numpy as np

np.random.seed(1000)

y = np.random.standard_normal(20)

x = np.arange(len(y))

plt.plot(x, y)
```

```python
plt.plot(y.cumsum(), 'b', lw=1.5) # objects can be passed with methods
plt.plot(y.cumsum(), 'ro')
plt.xlim(-1, 20) # directly set the minimum and maximum values of each axis
plt.ylim(np.min(y.cumsum()) - 1,
            np.max(y.cumsum()) + 1)
plt.xlabel('index')
plt.ylabel('A simple plot')
```

__Two dimensional datasets__

One can pass two-dimensional arrays to `plt.plot()`.

```python
y = np.random.standard_normal((20, 2)).cumsum(axis=0)

plt.figure(figsize=(10, 6))
plt.plot(y[:, 0], lw=1.5, label='1st') # data subsets are accessed separately, allowing you to attach individual labels
plt.plot(y[:, 1], lw=1.5, label='2nd')
plt.plot(y, 'ro')
plt.legend(loc=0) # 0 means best location
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
```

```python
y[:, 0] = y[:, 0] * 100 # create dataset that requires different scale

# Solution 1: use two y-axes (left and right) and the subplots overlay each other

fig, ax1 = plt.subplots() # defines the figure and the axis objects
plt.plot(y[:, 0], 'b', lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc=8)
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('A Simple Plot')
ax2 = ax1.twinx() # creates a second axis object that shares the x-axis
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'ro')
plt.legend(loc=0)
plt.ylabel('value 2nd')
```

```python
# Solution 2: create two separate sub-plots

plt.figure(figsize=(10, 6))
plt.subplot(211) # Define the upper subplot (numrows, numcols, fignum)
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc=0)
plt.ylabel('value')
plt.title('A Simple Plot')
plt.subplot(212) # Define the lower subplot
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'ro')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(121)
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('1st Data Set')
plt.subplot(122)
plt.bar(np.arange(len(y)), y[:, 1], width=0.5, # two different plot types
     color='g', label='2nd')
plt.legend(loc=0)
plt.xlabel('index')
plt.title('2nd Data Set')
```

```python
y = np.random.standard_normal((1000, 2))

plt.figure(figsize=(10, 6))
plt.scatter(y[:, 0], y[:, 1], marker='o')  # dedicated scatter plot function
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot')
```

```python
c = np.random.randint(0, 10, len(y))

plt.figure(figsize=(10, 6))
plt.scatter(y[:, 0], y[:, 1],
         c=c,
         cmap='coolwarm',
         marker='o')
plt.colorbar()
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot')
```

```python
plt.figure(figsize=(10, 6))
plt.hist(y, label=['1st', '2nd'], bins=25)
plt.legend(loc=0)
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram')
```
