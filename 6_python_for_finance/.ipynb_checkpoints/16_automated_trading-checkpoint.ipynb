### ML based trading strategy

Combination of ML with the historical data from the FXCM REST API to backtest the strategy. Uses the bid-ask spread as proportional transaction costs.

__Vectorised backtesting__
- Intraday data - in this case the five minute bars.
- Connect to FXCM ans retrieve five minute bar data for the month

ML strategy based on lagged return data that is binarized. Learns from historical patterns of upward and downward movements whether another upward or downward movement is likely.

```python
import fxcmpy
import math
import time
import numpy as np
import pandas as pd
import datetime as dt
#import cufflinks as cf
from pylab import plt

fxcmpy.__version__
plt.style.use('seaborn')
%matplotlib inline
```

```python
api = fxcmpy.fxcmpy(config_file='fxcm.cfg')
```

```python
data = api.get_candles('EUR/USD', period='m5',
                        start='2018-06-01 00:00:00',
                        stop='2018-06-30 00:00:00')
data.iloc[-5:, 4:]

#import pandas as pd
#data.to_csv (r'EUR-USD_5m_2018-06-01.csv', index=None, header=True)
#data=pd.read_csv('EUR-USD_5m_2018-06-01.csv')
```

```python
data.info()
```

```python
spread = (data['askclose'] - data['bidclose']).mean()
spread # average bid-ask spread
```

```python
data['midclose'] = (data['askclose'] + data['bidclose']) / 2
ptc = spread / data['midclose'].mean()
ptc # average proportional transaction costs given the average spread and the average mid close price
```

```python
data['midclose'].plot(figsize=(10, 6), legend=True)
```

```python
data['returns'] = np.log(data['midclose'] / data['midclose'].shift(1))
data.dropna(inplace=True)

lags=5
cols = []
for lag in range(1, lags + 1):
    col = 'lag_{}'.format(lag)
    data[col] = data['returns'].shift(lag) # lagged return data
    cols.append(col)
    
data.dropna(inplace=True)
data[cols] = np.where(data[cols] > 0, 1, 0) # transforms feature values into binary data
data['direction'] = np.where(data['returns'] > 0, 1, -1) # returns data to directional label data
data[cols + ['direction']].head()
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
# implement a support vector machine learning algorithm
# train and test the strategy based on a sequential train-test split
model = SVC(C=1, kernel='linear', gamma='auto')

split = int(len(data) * 0.80)

train = data.iloc[:split].copy()

model.fit(train[cols], train['direction'])
```

```python
accuracy_score(train['direction'], model.predict(train[cols]))
```

```python
test = data.iloc[split:].copy()
test['position'] = model.predict(test[cols])
accuracy_score(test['direction'], test['position'])
```

```python
sub = int(len(test)*0.1)

fig, ax1 = plt.subplots(figsize=(20, 12))

ax1.plot(test['position'].iloc[:sub])

ax2 = ax1.twinx()
ax2.plot(test['midclose'].iloc[:sub], 'r')

# far too many trades are taken. Chopped in and out of trades
```

```python
# next months data
data2 = api.get_candles('EUR/USD', period='m5',
                        start='2018-07-01 00:00:00',
                        stop='2018-07-31 00:00:00')
```

```python
data2['midclose'] = (data2['askclose'] + data2['bidclose']) / 2
data2['returns'] = np.log(data2['midclose'] / data2['midclose'].shift(1))
data2.dropna(inplace=True)

lags=5
cols = []
for lag in range(1, lags + 1):
    col = 'lag_{}'.format(lag)
    data2[col] = data2['returns'].shift(lag) # lagged return data
    cols.append(col)
    
data2.dropna(inplace=True)
data2[cols] = np.where(data2[cols] > 0, 1, 0) # transforms feature values into binary data
data2['direction'] = np.where(data2['returns'] > 0, 1, -1) # returns data to directional label data

data2['position'] = model.predict(data2[cols])
accuracy_score(data2['direction'], data2['position'])
```

Compare the performance of the strategy to the performance of the passive benchmark investment:

```python
# log returns for the ML strategy
test['strategy'] = test['position'] * test['returns']

# number of trades implied by the trading strategy
sum(test['position'].diff() != 0)
```

```python
test['strategy_tc'] = np.where(test['position'].diff() != 0,
                                test['strategy'] - ptc,
                                test['strategy'])
test[['returns', 'strategy', 'strategy_tc']].sum().apply(np.exp)
```

### Value-at-risk

It is quoted as a currency amount and represents the maximum loss to be expected given both a certain time horizon and a confidence level.

This method derives VaR based on log returns of the equity position. The time interval is fixed to a bar length of five minutes:

```python
import scipy.stats as scs
```

```python
percs = np.array([0.01, 0.1, 1.0, 2.5, 5.0, 10.0])

risk = pd.DataFrame(test['returns'])

#risk['returns'] = np.log(test['returns'] /
 #                        test['strategy_tc'].shift(1))
VaR = scs.scoreatpercentile(test['returns'], percs)
```

```python
def print_var():
    print('%16s %16s' % ('Confidence Level', 'Value-at-Risk'))
    print(33 * '-')
    for pair in zip(percs, VaR):
        print('%16.2f %16.3f' % (100 - pair[0], -pair[1]*10000))
```

```python
print_var()
```
